```python
data = [
["Sunny","Rich", "Yes", "Cinema"],
["Sunny",	"Rich",	"No",	"Tennis"],
["Windy",	"Rich",	"No",	"Cinema"],
["Rainy",	"Poor",	"Yes",	"Cinema"],
["Rainy",	"Rich",	"No",	"Stay-in"],
["Rainy",	"Poor",	"No",	"Cinema"],
["Windy",	"Poor",	"Yes",	"Cinema"],
["Windy",	"Rich",	"Yes",	"Stay-in"],
["Windy",	"Rich",	"No",	"Cinema"],
["Sunny",	"Rich",	"No",	"Tennis"],
["Sunny",	"Poor",	"Yes",	"Tennis"],
["Rainy",	"Rich",	"No",	"Stay-in"]];

import pandas as pd
df = pd.DataFrame(data)
df.columns = ['Weather', 	'Cash',	'Exam',	'Decision'];
print(df)
```

```python
df_x = pd.get_dummies(df[['Weather', 	'Cash',	'Exam']])
df_x.head()
```

```python
X= df_x.to_numpy()
print(X)
```

```python
y = df['Decision'].values.tolist()
print(y)
```

```python
'XXXXX ADD YOUR CODE HERE XXXXXXX'
'PRINT your name(s) and student ID(s), so the snapshot contains them'
'PRINT the tree, the predicted class, and the predicted probability'
from sklearn import tree
from sklearn import preprocessing
dtc = tree.DecisionTreeClassifier(criterion="entropy")
dtc.fit(X, y)
print("Names (ID): Huzaifa Mohammed (40242080), Mohammed Shurrab (40323793)," \
"\n Oleksandr Yasinovskyy (40241188) ")
tree.plot_tree(dtc)

X_test = [[0,1,0,1,0,1,0]]
y_pred = dtc.predict(X_test)
print("*no limit on tree-depth*")
print("Prediction: ", y_pred)
prob_y_pred = dtc.predict_proba(X_test)
print("Prediction Probability: ", prob_y_pred)
```
